# FOOTBALL PREMIER LEAGUE ANALYSIS OF 2017-18

# IMPORTING LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as pyp
```

# READING DATA

```python
football = pd.read_csv('football.csv')
```

# EXPLORATORY DATA ANALYSIS (EDA)

```python
football.describe()
```

```python
football.info()
```

# WORKING ON SESSION 2017-18

```python
working_data = football[football['Season']=='2017-18']
working_data
```

# TYPE CASTING

```python
working_data['Date'] = pd.to_datetime(working_data['Date'])
working_data['HTHG'] = working_data['HTHG'].astype('int64')
working_data['HTAG'] = working_data['HTAG'].astype('int64')
```

```python
working_data.info()
```

```python
working_data
```

# DATA FILTERING

```python
man_city = working_data[(working_data['HomeTeam']=='Man City') \
                        | (working_data['AwayTeam']=='Man City')]
man_city
```

```python
len(man_city)
```

```python
man_city_wins = working_data[((working_data['HomeTeam']=='Man City') & (working_data['FTR']=='H')) \
                        | ((working_data['AwayTeam']=='Man City') & (working_data['FTR']=='A'))]
man_city_wins
```

```python
len(man_city_wins)
```

```python
man_city_loss = working_data[((working_data['HomeTeam']=='Man City') & (working_data['FTR']=='A')) \
                        | ((working_data['AwayTeam']=='Man City') & (working_data['FTR']=='H'))]
man_city_loss
```

```python
len(man_city_loss)
```

```python
man_city_draw = working_data[((working_data['HomeTeam']=='Man City') & (working_data['FTR']=='D')) \
                        | ((working_data['AwayTeam']=='Man City') & (working_data['FTR']=='D'))]
man_city_draw
```

```python
len(man_city_draw)
```

# LOGICAL CONDITIONS WITH VISUALISATION

```python
frequency = [len(man_city_wins),len(man_city_loss),len(man_city_draw)]
grade = ['Win','Loss','Draw']
fig, ax = pyp.subplots(1,1, figsize = (6.5,7))
pyp.ylim(-3,38)
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.tick_params(bottom = False, left = False)
for location in ["left","right","bottom","top"]:
    ax.spines[location].set_visible(False)
ax.bar(grade,frequency,color=['lime','#FF0000','#C0C0C0'])
ax.text(0,-1.5,'Win',weight='bold')
ax.text(1,-1.5,'Loss',weight='bold')
ax.text(2,-1.5,'Draw',weight='bold')
ax.text(0.5,38,"MAN CITY (2017-18)",weight='bold')
ax.text(0,33,"32")
ax.text(1,3,"2")
ax.text(2,5,"4")
ax.axhline(y=36,xmin=0.1,xmax=0.9,linewidth=4,color='black')
xmin_h=0.1
xmax_h=0.12
for i in range(38):
    if (man_city.iloc[i][2]=='Man City') and (man_city.iloc[i][6]=='H'):
        ax.axhline(y=36,xmin=xmin_h,xmax=xmax_h,linewidth=4,color='green')
    elif (man_city.iloc[i][3]=='Man City') and (man_city.iloc[i][6]=='A'):
        ax.axhline(y=36,xmin=xmin_h,xmax=xmax_h,linewidth=4,color='green')
    elif (man_city.iloc[i][2]=='Man City') and (man_city.iloc[i][6]=='A'):
        ax.axhline(y=36,xmin=xmin_h,xmax=xmax_h,linewidth=4,color='red')
    elif (man_city.iloc[i][3]=='Man City') and (man_city.iloc[i][6]=='H'):
        ax.axhline(y=36,xmin=xmin_h,xmax=xmax_h,linewidth=4,color='red')
    elif (man_city.iloc[i][3]=='Man City') and (man_city.iloc[i][6]=='D'):
        ax.axhline(y=36,xmin=xmin_h,xmax=xmax_h,linewidth=4,color='grey')
    elif (man_city.iloc[i][2]=='Man City') and (man_city.iloc[i][6]=='D'):
        ax.axhline(y=36,xmin=xmin_h,xmax=xmax_h,linewidth=4,color='grey')
    xmin_h += 0.021
    xmax_h += 0.021
pyp.show()
```
