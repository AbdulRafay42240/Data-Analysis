# IMPORTING LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as pyp
```

# DECLARING MULTIPLE ROUTE FILES

```python
files = ['data_01.xlsx','data_02.xlsx','data_05.xlsx','data_06.xlsx','data_07.xlsx']
```

# INPUT FOR ROUTE

```python
num = int(input('NUM OF FILE: '))
num -= 1
```

# DATA READING

```python
def read(num):
    global data
    data = pd.read_excel(files[num])
    return data
read(num)
```

# REMOVING EXTRA COLUMNS

```python
def extra_cols(extras):
    global data
    data = data.drop(extras,axis=1)
    return data
extra_cols(['Mileage','Reference'])
```

```python
data.Status.unique()
```

# IGNITION ON OF CAR

```python
def ignite_on():
    global ignition_on
    ignition_on = data[data.Status == 'Ignition On']
    return ignition_on
ignite_on()
```

```python
len(ignition_on)
```

# IGNITION OFF OF CAR

```python
def ignite_off():
    global ignition_off
    ignition_off = data[data.Status == 'Ignition Off']
    return ignition_off
ignite_off()
```

```python
len(ignition_off)
```

# INPUT FOR ROUTE TRIP

```python
trip = int(input(f'Enter Trip Out of {len(ignition_on)}: '))
trip -= 1
```

# LOGIC TO GET THE REQUIRED DATAFRAME

```python
start = ignition_on.iloc[trip,:]
start
```

```python
end = ignition_off.iloc[trip,:]
end
```

```python
start_index = ignition_on.index
start_index
```

```python
end_index = ignition_off.index
end_index
```

```python
dup_data = data.iloc[start_index[trip]:end_index[trip]+1,:]
dup_data
```

# WHERE DRIVER PULL OFF SEAT BELT

```python
def seat_off(dup_data):
    global seat_off
    seat_off = dup_data[dup_data.Status == 'Driver SeatBelt Close']
    return seat_off
seat_off(dup_data)
```

```python
len(seat_off)
```

# WHERE DRIVER WORE SEAT BELT

```python
def seat_on(dup_data):
    global seat_on
    seat_on = dup_data[dup_data.Status == 'Driver SeatBelt Open']
    return seat_on
seat_on(dup_data)
```

```python
len(seat_on)
```

# WHERE DRIVER OVER SPEEDED

```python
def over_speed(dup_data):
    global over
    over = dup_data[dup_data['Speed(Km/h)'] > 80]
    return over
over_speed(dup_data)
```

# WHERE DRIVER HARSHLY BRAKED

```python
def harsh_brake(dup_data):
    global harsh_brake
    brake = dup_data.iloc[1:-1,2]
    brake = abs(brake.diff())
    brake = brake[brake>30]
    print(brake)
    harsh_brake = data.iloc[brake.index[0],:]
    return harsh_brake
harsh_brake(dup_data)
```

```python
pd.options.display.max_rows=None
```

# WHERE DRIVER HARSHLY ACCELERATED

```python
def harsh_acc(dup_data):
    global acc
    acc = dup_data.iloc[1:-1,2]
    acc = acc.diff()
    acc = acc[acc>20]
harsh_acc(dup_data)
```

# LOGIC TO GET THE EXACT DATAFRAME

```python
for i in acc.index:
    print(data.iloc[i,:])
```

# VISUALISATION ON DATAFRAMES

```python
pyp.plot(dup_data['Longitude'],dup_data['Latitude'],linestyle='dashed',color='green')
pyp.plot(over['Longitude'],over['Latitude'],linestyle='solid',color='red',linewidth=3)
pyp.scatter(seat_on['Longitude'],seat_on['Latitude'],color='black')
pyp.scatter(seat_off['Longitude'],seat_off['Latitude'],color='brown')
pyp.scatter(harsh_brake['Longitude'],harsh_brake['Latitude'],color='red')
for i in acc.index:
    harsh_acc = data.iloc[i,:]
    pyp.scatter(harsh_acc['Longitude'],harsh_acc['Latitude'],color='blue')

pyp.show()
```
