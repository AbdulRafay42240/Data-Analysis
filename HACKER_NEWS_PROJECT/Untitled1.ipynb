```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as pyp
```

```python
data = pd.read_csv('hacker_news.csv')
data
```

```python
data.head()
```

```python
ask_data = data[data['title'].str.contains('Ask HN:')].reset_index(drop=True)
ask_data
```

```python
show_data = data[data['title'].str.contains('Show HN:')].reset_index(drop=True)
show_data
```

```python
data = pd.concat([ask_data, show_data], ignore_index=True)
data
```

```python
sum(ask_data['num_comments'])
```

```python
sum(show_data['num_comments'])
```

```python
pyp.bar(['Avg Of Ask_HN','Avg Of Show_HN'],[sum(ask_data['num_comments'])/len(ask_data),\
        sum(show_data['num_comments'])/len(show_data)],width=0.7,color=['aqua','orange'],edgecolor='black')
pyp.xticks(rotation=90,weight='bold')
pyp.show()
```

```python
data['created_at'] = pd.to_datetime(data['created_at'])
```

```python
data.head()
```

```python
data['hrs'] = data['created_at'].dt.hour
```

```python
data.head()
```

```python
hold = data.pivot_table(values='num_comments',aggfunc='mean',index='hrs')
hold
```

```python
pyp.plot(['01 AM','02 AM','03 AM','04 AM','05 AM','06 AM','07 AM','08 AM','09 AM','10 AM',\
         '11 AM','12 AM','01 PM','02 PM','03 PM','04 PM','05 PM','06 PM','07 PM','08 PM','09 PM',\
         '10 PM','11 PM','12 PM'],hold['num_comments'],color='purple')
pyp.xticks(rotation=90,weight='bold')
pyp.yticks(weight='bold')
pyp.show()
```
